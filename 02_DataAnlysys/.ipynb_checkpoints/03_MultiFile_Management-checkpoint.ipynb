```python
file_counter = 0
for inputFile in fileList :
    # print(inputFile)
    # 반복실행중의 각각의 파일이름으로 파일 오푼
    with open(inputFile, 'r', newline='') as csv_in_file:
        filereader = csv.reader(csv_in_file)
        header = next(filereader)
        row_counter = 1
        # 헤더 이후의 데이터 들로 반복실행
        for row in filereader:
            row_counter = row_counter + 1
    # os.path.basename(input_file) : 파일명     
    # len(header): header 리스트의 갯수, 열의 갯수
    print('{0:s}: \t{1:d} rows \t{2:d} columns'.format(
                    os.path.basename(inputFile), row_counter, len(header)))
    file_counter += 1  # 파일 갯수 카운트
print('Number of files: {0:d}'.format(file_counter))
```

```python
# 멀티파일 합치기 - 파이썬
import csv
import glob
import os
# 첫번째 읽어온 파일인지에 대한 플래그 변수(표시)  
firstFile = True 
# 첫째 파일 읽은후 false로 값을 변경합니다
# 두번째부터 읽어온 파일에서는 맨첫줄을 지나치기 위해 값을 저장 & 활용하는 
# 변수입니다

fileList = glob.glob( os.path.join( 'C:/Users/JAVA01/Python/02_DataAnlysys' , \
                                  'sales_*.csv' ) )
for inputFile in fileList :
    with open(inputFile, 'r', newline='\n') as csv_in_file:
        # 파일 쓰기용 open 옵션
        # "w" : 새로 쓰기
        # "a" : 이어서 쓰기
        with open('salesMergePython.csv', 'w', newline='\n') as csv_out_file:
            filereader = csv.reader(csv_in_file)
            filewriter = csv.writer(csv_out_file)
            if firstFile : 
                for row in filereader:
                    filewriter.writerow(row)
                firstFile = False
            else : 
                header = next(filereader)
                for row in filereader:
                    filewriter.writerow(row)
```

```python
# 멀티 파일에서의 합계 평균 - 파이썬
import csv
import glob
import os

# 세개의 파일을 각각 sale_amount  에 대해 합계 평균을 내고 한개의 파일에 씁니다
output_header_list = ['file_name', 'total_sales', 'average_sales']

# with open('SumAverage.csv', 'a', newline='') as csv_out_file:
#    filewriter = csv.writer(csv_out_file)

# with 사용하지 않고 파일 오픈
csv_out_file = open('SumAverage.csv' , 'a', newline='')
filewriter = csv.writer(csv_out_file)

# 파일이름, 합계, 평균 열제목 출력
filewriter.writerow(output_header_list)  
print(output_header_list)
    
fileList = glob.glob( os.path.join( 'C:/Users/JAVA01/Python/02_DataAnlysys' , \
                                  'sales_*.csv' ) )
for inputFile in fileList :
    with open(inputFile, 'r', newline='\n') as csv_in_file:
        filereader = csv.reader(csv_in_file)
        
        # 각 파일별로 출력될 내용이 저장될 리스트를 생성
        output_list = []
        
        # 첫번째 출력내용 : 경로를 뺀 파일의 이름    output_list 에 append
        output_list.append( os.path.basename(inputFile) )
        
        # 합계와 갯수를 계산하여 저장할 변수들을  생성하고 0으로 초기화
        total_sales = 0.0
        number_of_sales = 0.0
        
        # 헤더를 읽어 내서 실제 데이터부터 읽어올 준비를 합니다
        header = next(filereader)
        
        # 라인단위의 반복 실행
        for row in filereader:
            # 합계 : Sale Amount 값에서 '$' 제거 ->  천단위구분기호(,) 제거  
            # -> float 형으로 변환 -> total_sales에 합산
            total_sales += float( row[3].strip('$').replace(',', '') )
            number_of_sales += 1.0   # 합산된 Coast 값의 갯수 카운트
        # 평균 계산후 소수점 두자리 조절된 문자로 저장
        average_sales = '{0:.2f}'.format( total_sales / number_of_sales )
        # 출력내용 리스트에 합계, 평균순으로 저장
        output_list.append(total_sales)
        output_list.append(average_sales)
        # 파일과 화면에 쓰고
        filewriter.writerow(output_list)
        print(output_list)
# with 를 사용하지 않았으므로 파일을 close 합니다
csv_out_file.close() 
```

```python
# 합계 평균 - 판다스
import pandas as pd
import glob
import os
fileList = glob.glob( os.path.join( 'C:/Users/JAVA01/Python/02_DataAnlysys' , \
                                  'sales_*.csv' ) )
all_data_frames = []
for file in fileList:
    data_frame = pd.read_csv(file)
    # print( data_frame.loc[:,'Sale Amount'] )
    k = []
    for value in data_frame.loc[:,'Sale Amount'] :
        # print(value)
        k.append( float( value.strip('$').replace(',' , '') ) )
    total_sales = pd.DataFrame( k ).sum()
    average_sales = pd.DataFrame( k ).mean()
    
    # 딕셔너리 형태로 파일이름과 합계 평균을 구성
    data = {'file_name': os.path.basename(file), \
            'total_sales': total_sales, 'average_sales': average_sales}
    # 생성되어 있는  all_data_frames 에 하나씩  append 합니다
    all_data_frames.append( pd.DataFrame(data,  \
                    columns=['file_name', 'total_sales', 'average_sales']) )
# all_data_frames 에 들어있는 데이터 들을 데이터프래임 형태로 연결결합
data_frames_concat = pd.concat(all_data_frames, axis=0, ignore_index=True)
data_frames_concat.to_csv('SumAveragePandas.csv', index = False)
print(data_frames_concat)
```

```python
import pandas as pd
import glob
import os
import csv
fileList = glob.glob( os.path.join( 'C:/Users/JAVA01/Python/02_DataAnlysys' , \
                                  'sales_*.csv' ) )
all_data_frames = []
# 개별 파일에 대한 합계 평균
for file in fileList:
    # 파일 오픈
    data_frame = pd.read_csv(file, index_col=None)
    # 합계 계산(압축 for 와 .sum() 사용)
    total_sales = pd.DataFrame([float(value.strip('$').replace(',','')) \
                        for value in data_frame.loc[ : , 'Sale Amount']]).sum()
    # 평균 계산(압축 for 와 .mean() 사용)
    average_sales = pd.DataFrame([float(value.strip('$').replace(',','')) \
                        for value in data_frame.loc[:, 'Sale Amount']]).mean()
    # 파일 이름과 계산된 값들로 딕셔너리 구성
    data = {'file_name': os.path.basename(file), \
            'total_sales': total_sales, 'average_sales': average_sales}
    # 딕셔너리 를 데이터 프레임으로 변환 -> all_data_frames 리스트에 append
    all_data_frames.append(  pd.DataFrame(data, \
                        columns=['file_name', 'total_sales', 'average_sales']) )
# all_data_frames  리스트를 데이터 프레임으로 변환(.concat() 사용)
data_frames_concat = pd.concat(all_data_frames, axis=0, ignore_index=True)
# 데이터 프레임으로 변환된 내용을 출력파일로 출력
data_frames_concat.to_csv('TotAvg.csv', index = False)
print(data_frames_concat)

# 파일 세개를 합치 합계 평균 
for file in fileList:
    data_frame = pd.read_csv(file, index_col=None)
    all_data_frames.append(data_frame)
data_frame_concat = pd.concat(all_data_frames, axis=0, ignore_index=True)
All_total = pd.DataFrame([float( str(value).strip('$').replace(',','') ) \
                for value in data_frame_concat.loc[ : , 'Sale Amount']]).sum()
All_average = pd.DataFrame( [float( str(value).strip('$').replace(',',''))\
                for value in data_frame_concat.loc[ : , 'Sale Amount']] ).mean()
output_header_list = ['file_name', 'All_total', 'All_average']
csv_out_file = open('TotAvg.csv' , 'a', newline='\n')
filewriter = csv.writer(csv_out_file)
filewriter.writerow('\n')  # 파일별 값들과 전체값들 사이 공백
print()
filewriter.writerow(output_header_list)
print(output_header_list)
output_data_list = ['sales_*.csv', All_total[0], All_average[0]]
filewriter.writerow(output_data_list)
print(output_data_list)
csv_out_file.close()
```
