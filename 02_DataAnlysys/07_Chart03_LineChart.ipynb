```python
import pandas as pd
marathon_2015_2017 = pd.read_csv("marathon_2015_2017.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
# 데이터 프레임에서  drop 을 사용하면 삭제할 컬럼을 선택하지만,
# columns 에 리스트로 컬럼명을 제시하면 골라내고자 하는 컬럼만 선택수집합니다
# 선택컬럼에 sort_values() 속성을 사용하여 특정 컬럼 기준 정렬된 데이터도
# 얻을수 있습니다

# 아래는 컬럼 선택 후 최종 기록으로 오름 차순 정렬
record = pd.DataFrame(marathon_2015_2017, columns=['5K','10K','15K','20K', \
            'Half','25K','30K','35K', '40K','Official Time']).sort_values( \
                                        by=['Official Time'], ascending=True)
record
```

```python
# insert 함수 : 데이커 프레임에 새롭게 컬럼을 추가하는 또하나의 방법
# .insert( 새로삽입되는 컬럼 위치, 컬럼 이름, 적용될 값)
record.insert(0, 'Rank', range(1, len(record)+1))
# 0 : 첫번째 열에 새로운 컬럼 삽입
# 'Rank' : 새로 생성되는 컬럼의 이름
# range(1, len(record)+1) : 1부터 1씩 증가하여  인원수 만큼 숫자 발생
record
```

```python
# rank  100 까지 필터링
top100 = record[ record['Rank']<=100 ]
top100
```

```python
# 한글 폰트 설정 & 폰트가 선명하게 보이도록 retina 설정
plt.rc("font", family="Malgun Gothic")
from IPython.display import set_matplotlib_formats
set_matplotlib_formats("retina")
```

```python
xData = top100['Rank']
yData_Off = top100['Official Time']
yData_10K = top100['10K'] 
yData_20K = top100['20K']
yData_30K = top100['30K']
```

```python
plt.figure(figsize=(20,10))
plt.xlabel("순위")
plt.ylabel("기록")
#  자주 사용하는 linestyle : '-', '--', '-.', ':'
#  자주 사용하는 marker : 'o', '*', '^' 등
plt.plot(xData, yData_Off, color='red', linestyle='--')
plt.plot(xData, yData_10K, color='blue', linestyle=':')
plt.plot(xData, yData_20K, color='green', marker='o')
plt.plot(xData, yData_30K, color='magenta', marker='^')
plt.show()
```
